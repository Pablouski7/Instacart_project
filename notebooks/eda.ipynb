```python
import pandas as pd

# Tiene dos estructuras principales
# Series = columna/vector
# Dataframe = tabla

serie = pd.Series([1, 2, 3, 4, 5])  
print(serie)
serie.name = 'nombre vector/columna'
print(serie)

df = pd.DataFrame({
    'columna1':[1,3,45,66,4],
    'columna2':['a','b','c','d', 'e']
})
print(df)
print(df.shape)

df = pd.read_csv('../data/products.csv', sep=';')
print(df.info())
print(df.columns)
print(df.shape)
print(df.index)
print(df.describe())
print(df.tail())
print(df.head())
columnas = ['id_producto', 'nombre_producto', 'id_isla', 'id_departamento']
df.columns = columnas
print(df.columns)

print(df['id_isla'].value_counts())
print(df.memory_usage().sum())
print(df.T)
print(df.sort_values('id_producto', ascending=False))
print(df[['id_isla', 'id_producto']])

print(df[:10])
print(df.iloc[[1,5], [1,2]]) # primera lista indices de las filas
                             # segunda lista indices de las columnas

print(df[df['id_departamento']>10])
print(df[(df['id_departamento']>10)&(df['id_departamento']<15)])
print(df['id_departamento']-df['id_isla'])
print(df.isna().sum())
print(df.isna().sum().sum())
print(df[df['nombre_producto'].isna()])
df['nombre_producto'] = df['nombre_producto'].fillna('Desconocido')
print(df['id_isla'].mean())
print(df['id_isla'].median())
print(df['id_isla'].apply(lambda x: -x))
print(df['nombre_producto'].apply(lambda x: x.lower()))
print(df.groupby('nombre_producto')['id_producto'].mean())
series_id_mean = df.groupby('nombre_producto')['id_producto'].mean()
series_id_mean.name = 'id_producto_mean'
df = df.join(series_id_mean, on=['nombre_producto'], how='inner')
print(df.head())

import matplotlib.pyplot as plt
df['id_producto_mean'].plot()
df['id_producto'].plot(kind='bar')
df.plot(figsize=[10,10])
```
